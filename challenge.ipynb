```python
from dotenv import load_dotenv
from litellm import completion 
load_dotenv(override=True)
from litellm.llms.anthropic.chat.transformation import AnthropicConfig
AnthropicConfig._is_opus_4_6_model = staticmethod(lambda m: any(v in m.lower() for v in ("opus-4-7")))
```

```python

prompt = """
What's the difference between the smallest twin primes whose sum is divisible by 31? Answer only with the difference as quickly as you can.
"""
```

```python
messages = [{"role": "user", "content": prompt}]
```

```python
result = completion(messages=messages, model="gpt-5.5", reasoning_effort="xhigh")
print(result.choices[0].message.content)
```

```python
result = completion(messages=messages, model="gemini/gemini-3.1-pro-preview")
print(result.choices[0].message.content)
```

```python
result = completion(messages=messages, model="gpt-5", reasoning_effort="high")
print(result.choices[0].message.content)
```

```python


result = completion(model="anthropic/claude-opus-4-7", messages=messages, max_tokens=100, thinking={"type": "adaptive"}, output_config={"effort": "max"})
print(result.choices[0].message.content)
```
